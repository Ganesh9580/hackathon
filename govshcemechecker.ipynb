```python
import streamlit as st
import pandas as pd

# -------------------------------
# Load dataset
# -------------------------------
@st.cache_data
def load_data():
    df = pd.read_csv("gov_schemes.csv")
    df["max_income"] = df["max_income"].replace("any", 99999999).astype(int)
    return df

df = load_data()

# -------------------------------
# UI
# -------------------------------
st.set_page_config(page_title="Gov Scheme Checker", layout="centered")

st.title("üáÆüá≥ Government Scheme Eligibility Checker")
st.write("Enter your details to get recommended government schemes")

age = st.number_input("Age", min_value=0, max_value=120, value=25)
income = st.number_input("Annual Income (‚Çπ)", min_value=0, value=200000)
state = st.selectbox("State", sorted(df["state"].unique()))
occupation = st.selectbox("Occupation", sorted(df["occupation"].unique()))

# -------------------------------
# Matching logic
# -------------------------------
def check_eligibility(df, age, income, state, occupation):
    return df[
        (df["min_age"] <= age) &
        (df["max_age"] >= age) &
        (df["max_income"] >= income) &
        (
            (df["occupation"] == occupation) |
            (df["occupation"] == "any")
        ) &
        (
            (df["state"] == state) |
            (df["state"] == "any")
        )
    ]

# -------------------------------
# Button
# -------------------------------
if st.button("üîç Check Eligible Schemes"):
    result = check_eligibility(df, age, income, state, occupation)

    if len(result) == 0:
        st.error("‚ùå No scheme found for your details")
    else:
        st.success(f"‚úÖ You are eligible for {len(result)} schemes")

        for _, row in result.iterrows():
            st.markdown(f"### {row['scheme_name']}")
            st.write(row["description"])
            st.markdown("---")
```
